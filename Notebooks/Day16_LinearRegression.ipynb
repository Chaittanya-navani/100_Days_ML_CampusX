```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("real_placement.csv")
```

```python
df
```

```python
plt.scatter(df["cgpa"], df["package"])
plt.xlabel("CGPA")
plt.ylabel("Package")
```

```python
X = df.iloc[:,0:1]
y = df.iloc[:,-1]
y
```

```python

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
X_test
```

```python
y_test
```

```python
lr.predict(X_test.iloc[0].values.reshape(1, 1))
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(X_train,lr.predict(X_train),color='red')
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```

```python
m = lr.coef_ # This is slope (m)
m
```

```python
b = lr.intercept_
b
```

```python
# Manualy predicting values cuz we know the mathematical intuition of Linear regression
m * 7.4 + b
```
