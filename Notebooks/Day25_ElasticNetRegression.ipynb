```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
X, y = load_diabetes(return_X_y = True) 
```

```python
X.shape
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 2)
```

```python
X_train.shape
```

```python
y_train.shape
```

***Comparing Simple Linear, Ridge and Lasso with R2_Score***

```python
#Linear Regression 
lin = LinearRegression()
lin.fit(X_train, y_train)
y_pred1 = lin.predict(X_test)
r2_score(y_test, y_pred1)
```

```python
#Ridge Regularization
rid = Ridge(alpha=0.005)
rid.fit(X_train, y_train)
y_pred2 = rid.predict(X_test)
r2_score(y_test, y_pred2)
```

```python
#Lasso regularization
las = Lasso(alpha=0.005)
las.fit(X_train, y_train)
y_pred3 = las.predict(X_test)
r2_score(y_test, y_pred3)
```

***Now with Elastic Net Regression***

```python
elasn = ElasticNet(alpha=0.005, l1_ratio=0.9)
elasn.fit(X_train, y_train)
y_pred4 = elasn.predict(X_test)
r2_score(y_test, y_pred4)
```
