```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0, n_classes=2, n_clusters_per_class=1, random_state=41, hypercube=False, class_sep=20)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap="winter", s=80)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression(penalty=None, solver='sag')
log_reg.fit(X, y)
```

```python
log_reg.coef_
```

```python
log_reg.intercept_
```

```python
m1 = -(log_reg.coef_[0][0]/log_reg.coef_[0][1])
b1 = -(log_reg.intercept_[0]/log_reg.coef_[0][1])
```

```python
x_input = np.linspace(X[:, 0].min(), X[:, 0].max(), 100)
y_input = m1*x_input + b1
```

```python
plt.figure(figsize=(12, 6))
plt.plot(x_input, y_input, color='red', linewidth = 3)
plt.scatter(X[:, 0], X[:, 1], c = y, cmap='winter', s= 100)
plt.ylim(-3, 2)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Graph 1 - Without ylim
axes[0].plot(x_input, y_input, color='red', linewidth=3)
axes[0].scatter(X[:,0], X[:,1], c=y, cmap='winter', s=100)
axes[0].set_title('Without ylim')

# Graph 2 - ylim(-3, 2)
axes[1].plot(x_input, y_input, color='red', linewidth=3)
axes[1].scatter(X[:,0], X[:,1], c=y, cmap='winter', s=100)
axes[1].set_ylim(-3, 2)
axes[1].set_title('ylim(-3, 2)')

# Graph 3 - ylim(-1, 0)
axes[2].plot(x_input, y_input, color='red', linewidth=3)
axes[2].scatter(X[:,0], X[:,1], c=y, cmap='winter', s=100)
axes[2].set_ylim(-1, 0)
axes[2].set_title('ylim(-1, 0) - Bahut zoom in')

plt.show()
```
